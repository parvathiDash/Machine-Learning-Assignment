```python
import pandas as pd
import numpy as np
```

# Given Dataframe

```python
obj = {'From_To':['LoNDon_paris','MAdrid_miLAn','londON_StockhOlm','Budapest_PaRis','Brussels_londOn'],
            'FlightNumber':[10045,np.nan,10065,np.nan,10085],
            'RecentDelays':[[23,47],[],[24,43,87,45],[13],[67,32]],
                  'Airline': ['KLM(!)','<Air France(12)','(British Airways. )','12.Air France','Swiss Air']
}

```

```python
df = pd.DataFrame(obj)
df
```

# Adding missing values to the Flight Number Column

```python
for i in range(1,df['FlightNumber'].count()+1):
    if pd.isnull(df.loc[i,'FlightNumber']):
        df.loc[i,'FlightNumber'] = df.loc[i-1,'FlightNumber']+10
```

# Flight Number column converted to int

```python
df['FlightNumber'] = df['FlightNumber'].astype(int)
```

```python
df
```

# Creating a new dataframe from another dataframe using Copy() method without affecting the original dataframe:

```python
df_New_Temp = df.copy()
```

```python
df_New_Temp
```

```python
df_New_Temp['From_To'] = df_New_Temp['From_To'].str.split('_')
df_New_Temp
```

```python
df
```

```python
df_New_Temp['From'] = df_New_Temp['From_To'].str[0]
df_New_Temp
```

```python
df_New_Temp['To'] = df_New_Temp['From_To'].str[1]
df_New_Temp
```

```python
df_New_Temp['From'] = df_New_Temp['From'].str.capitalize()
df_New_Temp
```

```python
df_New_Temp['To'] = df_New_Temp['To'].str.capitalize()
df_New_Temp
```

```python
df_New_Temp.drop(['From_To','FlightNumber','RecentDelays','Airline'],axis = 1,inplace = True)
```

```python
df_New_Temp
```

```python
df = pd.concat([df_New_Temp,df],axis = 1)
df
```

```python
df.drop('From_To',axis = 1,inplace = True)
```

```python
df
```

```python
d2 = (pd.DataFrame(df['RecentDelays'].values.tolist(), index=df.index).rename(columns = lambda x: 'Delay_{}'.format(x+1)))
d2
```
