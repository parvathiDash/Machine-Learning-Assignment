# In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
from sklearn import decomposition
from sklearn import datasets
```

# loading iris datasets using load_iris

```python
iris = datasets.load_iris()
```

#  Axes3D for plotting 3D figure and decomposition for reducing the data availability to 3 components and used scatter for plotting the 3D plot.. where normal matplotlib scatter gives only 2D scatter

```python
X = iris.data
y=iris.target
fig = plt.figure()
ax =Axes3D(fig,elev=-150, azim=110,proj_type='ortho')
ax
# decomposition for reducing the components
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)

ax.scatter(X[:,0],X[:,1],X[:,2],c=y,s=40)
ax.set_xlabel('x axis')
#ax.xaxis.set_ticklabels([])
ax.set_ylabel('y axis')
#ax.yaxis.set_ticklabels([])
ax.set_zlabel('z axis')
#ax.zaxis.set_ticklabels([])
plt.show()
```
